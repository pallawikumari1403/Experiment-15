Part A : Data Normalization and Data Type Conversion


```python
# Dataset
import pandas as pd
import numpy as n
data = {
    'Product':['Laptop','Mobile','Tablet','Headphones','Smartwatch','Camera'],
    'Price' : [55000,20000,30000,2000,8000,45000],
    'Units_sold':[25,60,35,80,50,15],
    'Discount':[10,15,12,20,18,8]
}
df = pd.DataFrame(data)
print(df)
```

```python
#1. Min-Max Normalizatiion
# Min-Max Normalization

df['Price_Minmax'] = (df['Price'] - df['Price'].min()) / (df['Price'].max()
                            - df['Price'].min())
print(df[['Product','Price','Price_Minmax']])
```

```python
# Normalization of Multiple Columns
cols = ['Price','Units_sold','Discount']
df_norm = (df[cols]-df[cols].min())/(df[cols].max()-df[cols].min())
print(df_norm)
```

```python
#2. Z-Score Normalization
df['Units_Zscore'] = (df['Units_sold'] - df['Units_sold'].mean())/df['Units_sold'].std()
print(df[['Product','Units_sold','Units_Zscore']])
```

```python
zcols = ['Price','Units_sold','Discount']
df_Zscore = (df[zcols]-df[zcols].mean())/df[zcols].std()
print(df_Zscore)
```

```python
#3. Decimal Scaling
df['Price_Decimal'] = df['Price']/100000
print(df[['Product','Price','Price_Decimal']])
```

```python
dcols = ['Price','Units_sold','Discount']
df_colscale = df[dcols]/1000
print(df_colscale)
```

Part B: Turning categorical variables into Quantitative Variables in python

```python
import pandas as pd
import numpy as np
data = {
'Order_ID':[101,102,103,104,105,106],
'Customer_Gender':['Male','Female','Female','Male','Female','Male'],
'Payment_Method':['UPI','Credit Card','Debit Card','COD','UPI','Debit Card'],
'Product_Category':['Electronics','Clothing','Home','Electronics','Beauty','Electronics'],
'City':['Pune','Mumbai','Delhi','Bangalore','Pune','Hyderabad'],
'Order_Value':[4500,2300,3200,5200,1800,4100]
}
df = pd.DataFrame(data)
print(df)
```

```python
# Method 1:Label Encoding
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['Gender_Label'] = le.fit_transform(df['Customer_Gender'])
df['City_Label'] = le.fit_transform(df['City'])
df['Payment_Label'] = le.fit_transform(df['Payment_Method'])
df['Product_Label'] = le.fit_transform(df['Product_Category'])
print(df)
```

```python
# Method 2: One Hot Encoding
df_encoded = pd.get_dummies(df,columns=['Payment_Method'])
print(df_encoded)
```

```python
# Method 3: Dummy Encoding
df_dummy = pd.get_dummies(df,columns=['Payment_Method'],drop_first=True)
print(df_dummy)
```
